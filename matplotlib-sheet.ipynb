```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
pincode = pd.read_csv("C:/Users/Ayush Jaiswal/Downloads/Pincode_30052019.csv", encoding= "ISO-8859-1", engine= 'python')
pincode["StateName"]
ax.set_title("Indian Pincode Analysis", color = "#00ffff")
ax.set_xlabel("State Name", color = "purple")
ax.set_ylabel("No. Of Pincode", color = "purple")
state_name = pincode.groupby("StateName")["Pincode"].count()
print(state_name.head(5))
state_name.plot(kind = "bar", color = "#ff80ff")
ax.plot(kind = "bar")
plt.show()
```

```python
group1 = pincode.groupby("StateName")["District"].count()
print(group1)
group1.plot(kind = 'scatter', x = 'StateName', y = 'District', color = 'cyan')
```
